```python
import numpy as np
import matplotlib.pyplot as plt 
import os
import scipy.misc as sm

def rgb2gray(rgb):

    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    gray = 0.2989 * r + 0.5870 * g + 0.1140 * b

    return gray


def visualize(img, format=None, gray=False):
    plt.imshow(img, format)
    plt.show()
```

```python
import scipy.ndimage
from keras.preprocessing import image
import numpy
import cv2
import numpy as np
from matplotlib import pyplot
import os
import canny_edge_detector as ced

image_size = (196, 196)


path = "./images/101_ObjectCategories"
dirs = os.listdir(path)
i = 0
for dir in dirs:
    files = os.listdir(path+"/"+dir)
    j = 0
    i+=1
    for file in files:
        j+=1
        img = image.load_img(path+"/"+dir+"/"+file, dir, target_size=image_size)
        img_data = image.img_to_array(img)
        img_gray = rgb2gray(img_data)
        detector = ced.cannyEdgeDetector(
            img_gray, 
            sigma=1.4,
            kernel_size=5,
            lowthreshold=0.09, 
            highthreshold=0.17,
            weak_pixel=100
        )
        img_final = detector.detect()
        img_final = numpy.repeat(img_final, 3)
        img_final.shape = (image_size[0], image_size[1], 3)
        if not os.path.exists("./images/unprocessed"+"/"+dir):
            os.mkdir("./images/unprocessed"+"/"+dir)
        image.save_img("./images/unprocessed"+"/"+dir+"/"+file, img_final)
        print("Processed: {0}/{1} - {2}/{3}".format(i, len(dirs), j, len(files)))
#         visualize(img_final, 'gray')
```
